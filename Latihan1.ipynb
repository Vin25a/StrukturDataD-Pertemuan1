5 Nama Mahasiswa

```python
Mahasiswa = ["Andi", "Budi", "Citra", "Dewi", "Eko"]
print("List Nama Mahasiswa:")
print(Mahasiswa)
```

Menambahkan Element Append

```python
Mahasiswa.append("Fajar")
print("\nSetelah Menambahkan Nama Baru:")
print(Mahasiswa)
```

Hapus Satu Element Menggunakan Remove

```python
Mahasiswa.remove("Andi")
print("\nSetelah Remove Nama:")
print(Mahasiswa)
```
